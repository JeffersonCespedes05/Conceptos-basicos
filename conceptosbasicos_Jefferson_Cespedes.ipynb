Conceptos basicos

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a sample DataFrame with 'ventas' and 'publicidad' with 1000 data points
data = {'publicidad': np.random.rand(1000) * 100,
        'ventas': np.random.rand(1000) * 500 + 100}
df = pd.DataFrame(data)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['publicidad'], df['ventas'])
plt.title('Gráfica de Dispersión entre Ventas y Publicidad')
plt.xlabel('Publicidad')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```

Crear un programa donde se solicita ingresar su nombre y su edad, que automaticamente imprima un mensaje saludandolo y mencionando cuantos años tendra en 5 años mas

```python
# Solicitar nombre y edad al usuario
nombre = input("Por favor, ingresa tu nombre: ")
edad_str = input("Por favor, ingresa tu edad: ")

# Convertir la edad a un número entero
edad = int(edad_str)

# Calcular la edad en 5 años
edad_en_5_años = edad + 5

# Imprimir el mensaje de saludo y la edad en 5 años
print(f"Hola, {nombre}. En 5 años tendrás {edad_en_5_años} años.")
```

```python
def areacirculo(radio):
  pi_value = 3.14159
  area = pi_value * radio ** 2
  return area
```

```python
#ejemplo de celsius a F

def celsius_a_fahrenheit(celsius):
  fahrenheit = (celsius * 9/5) + 32
  return fahrenheit
```

```python
celsius_a_fahrenheit (25)
```
